```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
df.info()
```

Note that there are NaN values in horsepower. NaN values need to be dropped when using sns.pairplot(). Otherwise, the line would throw an error: max must be larger than min in range parameter.

```python
sns.pairplot(df[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']].dropna())
plt.show()
```

```python
sns.pairplot(df[['mpg', 'horsepower', 'weight', 'origin']].dropna(), hue='origin', kind='reg')
plt.show()
```

```python
sns.lmplot(x='horsepower', y='mpg', data=df)
plt.show()
```

```python
sns.lmplot(x='horsepower', y='mpg', hue='origin', data=df)
plt.show()
```

```python
sns.lmplot(x='horsepower', y='mpg', col='origin', data=df)
plt.show()
```

```python
sns.residplot(x='horsepower', y='mpg', data=df)
plt.show()
```

```python
plt.scatter(df['horsepower'], df['mpg'], color='gray', label='data points', s=8)
sns.regplot(x='horsepower', y='mpg', data=df, scatter=None, label='order 1')
sns.regplot(x='horsepower', y='mpg', data=df, order=2, scatter=None, color='green', label='order 2')
plt.show()
```

```python
sns.stripplot(x='origin', y='mpg', data=df, s=4, jitter=True)
plt.show()
```

```python
sns.swarmplot(x='origin', y='mpg', hue='cylinders', data=df, s=4)
plt.legend(loc='lower right', ncol=5)
plt.show()
```

```python
sns.violinplot(x='origin', y='mpg', data=df)
plt.show()
```

```python
sns.jointplot(x='horsepower', y='mpg', data=df, kind='reg')
sns.jointplot(x='horsepower', y='mpg', data=df, kind='kde')
plt.show()
```
