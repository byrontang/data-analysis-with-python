## Random walk simulation

```python
repeat = 10
steps = 100
paths = []
for i in range(repeat):
    path = [0]
    for i in range(steps):
        toss = np.random.randint(0,2)
        if toss == 0:
            path.append(path[i-1] - 1)
        if toss == 1:
            path.append(path[i-1] + 1)
    paths.append(path)
#print(paths)
```

```python
for i in range(10):
    plt.plot(paths[i])
plt.show()
```

In random walk, the variance of Xi grows with i.

```python
plt.plot(np.transpose(paths))
plt.show()
```
