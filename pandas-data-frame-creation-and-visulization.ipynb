#### Edit x axis and figure size

```python
df.index = df[['year', 'month']]
df.passengers.plot()
plt.xticks(np.arange(len(df)+1, step = 3), df.index[::3], rotation=45)
#plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [20, 5]
plt.show()
```

#### zoom in

```python
plt.clf()
df.index = df[['year', 'month']]
df.passengers.plot()
plt.xticks(np.arange(len(df)+1, step = 3), df.index[::3], rotation=45)
plt.axis([120, 144, 300, 650])
plt.show()
```

#### pdf and cdf

```python
tips = sns.load_dataset('tips')
tips['fraction'] = tips.tip/tips.total_bill
tips.head()
```

```python
# This formats the plots such that they appear on separate rows
fig, axes = plt.subplots(nrows=2, ncols=1)

# Plot the PDF
tips.fraction.plot(ax=axes[0], kind='hist', bins=30, normed=True, range=(0,.3))

# Plot the CDF
tips.fraction.plot(ax=axes[1], kind='hist', bins=30, normed=True, cumulative=True, range=(0,.3))
plt.show()
```
